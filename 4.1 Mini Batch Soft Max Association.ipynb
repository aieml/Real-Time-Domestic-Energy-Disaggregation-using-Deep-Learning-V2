```python
import numpy as np
import os
import pandas as pd

data_path='data'
ref_clusters=[3.68231636,173.4372239,429.99911234,129.58479891]
```

```python
min_batch_soft=[]
count=0

min_files=os.listdir(os.path.join(data_path,'MINI_BATCHES_AVG'))
for file in min_files:
    
    if file.endswith(".npy"):
        
        channel_path=os.path.join(data_path,'MINI_BATCHES_AVG',file)
        dataset = np.load(channel_path)
        main=dataset[:,0]
        appliance=dataset[:,1]
        for i in range(len(appliance)):
            total=np.sum(np.array([1/np.power(ref_clusters[j]-appliance[i],2) for j in range (len(ref_clusters))]))
            prob0=(1/np.power(ref_clusters[0]-appliance[i],2))/total
            prob1=(1/np.power(ref_clusters[1]-appliance[i],2))/total
            prob2=(1/np.power(ref_clusters[2]-appliance[i],2))/total
            prob3=(1/np.power(ref_clusters[3]-appliance[i],2))/total
            #min_batch_soft.append([main[i],[prob0,prob1,prob2,prob3]])
            #print(main[i],prob0,prob1,prob2,prob3)
            min_batch_soft.append(1)
            
        print(len(min_batch_soft))   
        channel_save_path=os.path.join(data_path,'SOFT_ASSOCIATION','mini_batch'+str(count)+'.dat')
        np.save(channel_save_path,min_batch_soft)
        #np.savetxt(channel_save_path,min_batch_soft)
        min_batch_soft=[]
        print(len(min_batch_soft))
        count=count+1
```
