```python
import pandas as pd
import numpy as np
import os
```

```python
def forward_fill(dataset,house,channel):
    dataset=dataset[dataset[:, 0].argsort()]
    time=np.array(dataset[:,0],dtype=np.int)
    data=dataset[:,1]
    count=0
    new_time=[]
    new_data=[]

    for i in range(len(data)-1):
        if(time[i+1]!=time[i]+1):
            count=count+1
            gap=time[i+1]-time[i]
            if(gap<20):
                for j in range(gap):
                    new_time.append(time[i]+j)
                    new_data.append(data[i])
            else:
                print('GAP>20',house,channel,i)
                new_time.append(time[i])
                new_data.append(data[i])                 
        else:
            new_time.append(time[i])
            new_data.append(data[i])            

    new_time.append(time[len(data)-1])
    new_data.append(data[len(data)-1])
    
    new_dataset=[]
    
    for i in range(len(new_data)):
        new_dataset.append([new_time[i],new_data[i]])
    return np.array(new_dataset)
```

```python
path='low_freq'
data_path='data'
houses=['house_1','house_2','house_3','house_4','house_6']
main={'house_1':'channel_2.dat','house_2':'channel_2.dat','house_3':'channel_2.dat','house_4':'channel_2.dat','house_5':'channel_2.dat','house_6':'channel_2.dat'}
dishwasher={'house_1':'channel_6.dat','house_2':'channel_10.dat','house_3':'channel_9.dat','house_4':'channel_15.dat','house_5':'channel_20.dat','house_6':'None'}
microwave={'house_1':'channel_11.dat','house_2':'channel_6.dat','house_3':'channel_16.dat','house_4':'None','house_5':'channel_3.dat','house_6':'None'}
refrigerator={'house_1':'channel_5.dat','house_2':'channel_9.dat','house_3':'channel_7.dat','house_4':'None','house_5':'channel_18.dat','house_6':'channel_8.dat'}
appliances=[main,dishwasher,microwave,refrigerator]
```

```python
for house in houses:
    for appliance in appliances:
        channel=appliance[house]
        if(channel!='None'):
            channel_path=os.path.join(path,house,channel)
            dataset = pd.read_csv(channel_path, sep="\s+", usecols=[0,1],header=None).values
            new_dataset=forward_fill(dataset,house,channel)
            channel_save_path=os.path.join(data_path,'FORWARD_FILLED_20S',house,channel)
            np.save(channel_save_path,new_dataset)
            np.savetxt(channel_save_path,new_dataset)
```
